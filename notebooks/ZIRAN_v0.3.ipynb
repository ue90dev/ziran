```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

print("Loading ZIRAN's AI brain... ⏳")

model_name = "microsoft/Phi-3-mini-4k-instruct"

tokenizer = AutoTokenizer.from_pretrained(
    model_name,
    trust_remote_code=True
)

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float16,
    device_map="auto",
    trust_remote_code=True,
    attn_implementation="eager"
)

print("=" * 45)
print("   ✅ ZIRAN's AI Brain Loaded!")
print("   Model : Phi-3 Mini")
print("   Device: GPU ✅")
print("=" * 45)
```

```python
# ================================
# ZIRAN v0.3 - AI Powered Chat
# Built in Bangladesh 🇧🇩
# ================================

def ziran_think(user_input):
    # Format message for Phi-3
    messages = [
        {
            "role": "system",
            "content": """You are ZIRAN, the first Bangladeshi AI assistant.
You were built in Bangladesh 🇧🇩 by a young Bangladeshi developer.
You are helpful, friendly and proud of Bangladesh.
Keep answers short and clear."""
        },
        {
            "role": "user",
            "content": user_input
        }
    ]

    # Convert to model format
    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )

    # Tokenize input
    inputs = tokenizer(text, return_tensors="pt").to(model.device)

    # Generate response
    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=200,
            temperature=0.7,
            do_sample=True,
            pad_token_id=tokenizer.eos_token_id
        )

    # Decode response
    response = tokenizer.decode(
        outputs[0][inputs['input_ids'].shape[1]:],
        skip_special_tokens=True
    )

    return response.strip()

# ================================
# ZIRAN Chat Loop
# ================================
print("=" * 45)
print("   ◈  ZIRAN v0.3  ◈")
print("   First Bangladeshi AI 🇧🇩")
print("   Powered by Phi-3 Mini")
print("=" * 45)
print("   Type 'bye' to exit")
print("=" * 45)
print()

while True:
    user = input("You: ")

    if user.lower() in ["bye", "exit"]:
        print("ZIRAN: Goodbye! 🇧🇩")
        break

    print("ZIRAN: thinking... ⏳")
    response = ziran_think(user)
    print(f"ZIRAN: {response}")
    print()
```
