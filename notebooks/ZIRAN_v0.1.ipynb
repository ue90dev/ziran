```python
# ================================
# ZIRAN v0.1 - First Chatbot
# Built in Bangladesh 🇧🇩
# ================================

# ZIRAN's knowledge base
responses = {
    "hello": "Hello! I am ZIRAN 🇧🇩",
    "hi": "Hi there! ZIRAN here!",
    "name": "My name is ZIRAN!",
    "who": "I am the first Bangladeshi AI!",
    "country": "I was born in Bangladesh 🇧🇩",
    "version": "I am ZIRAN version 0.1",
    "bye": "Goodbye! Come back soon! 🇧🇩",
    "help": "I can answer basic questions for now!"
}

# ZIRAN's brain function
def ziran_think(user_input):
    # Convert to lowercase
    user_input = user_input.lower()

    # Check each keyword
    for keyword in responses:
        if keyword in user_input:
            return responses[keyword]

    # Default response
    return "I am still learning... Ask me something else!"

# ================================
# ZIRAN Chat Loop
# ================================
print("=" * 40)
print("   Welcome to ZIRAN v0.1 🇧🇩")
print("   Type 'bye' to exit")
print("=" * 40)

while True:
    # Get user input
    user = input("You: ")

    # Exit condition
    if user.lower() == "bye":
        print("ZIRAN: Goodbye! 🇧🇩")
        break

    # Get ZIRAN's response
    response = ziran_think(user)
    print(f"ZIRAN: {response}")
    print()
```
